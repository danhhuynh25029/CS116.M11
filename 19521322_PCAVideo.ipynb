```python
import cv2 
import numpy as np
from sklearn.decomposition import PCA
from google.colab.patches import cv2_imshow
v_matrix = []
pca = PCA(n_components=10)
cap = cv2.VideoCapture("/content/video1.mp4")
frame_width  = 400
frame_height = 350       
out = cv2.VideoWriter('output.mp4',cv2.VideoWriter_fourcc('M','J','P','G'), 30, (frame_width,frame_height),0)
while cap.isOpened():
  ret,frame = cap.read()
  if ret == True:
    frame = cv2.resize(frame,(frame_width,frame_height))
    frame  = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    pca.fit(frame)
    frame_reduced = pca.transform(frame)
    frame_recovered = pca.inverse_transform(frame_reduced)
    frame_recovered = np.uint8(frame_recovered)
    out.write(frame_recovered)
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
  else:
    break
cap.release()
out.release()
cv2.destroyAllWindows()
```
