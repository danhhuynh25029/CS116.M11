```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
df = gpd.read_file('data/TH1/population_shapefile/Population_District_Level.shp')
df.plot(figsize=(6, 6))
plt.show()
```

```python
df.head()
```

```python
df = df.to_crs(epsg=4326)
df.crs
```

```python
m = folium.Map(location=[40.70, -73.94], zoom_start=10, tiles='CartoDB positron')
m
```

```python
m = folium.Map(location=[10.8, 107], zoom_start=10, tiles='CartoDB positron')
for _, fr in df.iterrows():
    g = gpd.GeoSeries(fr['geometry']).simplify(tolerance=0.001).to_json()
    g = folium.GeoJson(data=g, style_function=lambda _: dict({'fillColor': 'yellow'}))
    folium.Popup(fr['Dist_Name']).add_to(g)
    g.add_to(m)
m
```
