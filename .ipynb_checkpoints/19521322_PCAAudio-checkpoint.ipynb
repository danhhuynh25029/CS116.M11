```python
from scipy.io import wavfile
import IPython
from sklearn.decomposition import PCA
import numpy as np
```

```python
sr, audio = wavfile.read('/content/file_example_WAV_1MG.wav')
```

```python
IPython.display.Audio("/content/file_example_WAV_1MG.wav")
```

```python
print(len(audio))
arr = []
for i in range(0,audio.shape[0]):
  for j in range(1,audio.shape[1]):
    arr.append(audio[i][j])
arr = np.array(arr)
arr
```

```python
z = 512 - len(audio)%512
arr_zero = np.zeros(z).astype('int16')
arr_zero
```

```python
arr_zero = np.array(arr_zero)
full_arr = np.concatenate([arr,arr_zero])
```

```python
arr_reshape = np.reshape(full_arr,(-1,512))
```

```python
pca = PCA(n_components = 100)
pca.fit(arr_reshape)

# Giam so chieu
audio_reduced = pca.transform(arr_reshape)
# Khoi phuc
audio_recovered = pca.inverse_transform(audio_reduced)
```
