```python
import geopandas as gpd
data = gpd.read_file("data/TH1/population_shapefile/Population_Ward_Level.shp")
data.head()
```

```python
m = max(data["Shape_Leng"])
mi = min(data["Shape_Leng"])
max_area = data[m == data['Shape_Leng']]
max_pop2019 = data[max(data['Pop_2019']) == data['Pop_2019']]
min_area = data[mi == data["Shape_Leng"]]
min_pop2019 = data[min(data['Pop_2019']) == data['Pop_2019']]
print("Phuong co dien tich lon nhat : {} {}".format(max_area['Com_Name'].values[0],max_area['Dist_Name'].values[0]))
print("Phuong co dan so lon nhat : {} {}".format(max_pop2019['Com_Name'].values[0],max_area['Dist_Name'].values[0]))
print("Phuong co dien tich thap nhat : {} {}".format(min_area['Com_Name'].values[0],min_area['Dist_Name'].values[0]))
print("Phuong co dan so thap nhat : {} {}".format(min_pop2019['Com_Name'].values[0],min_pop2019['Dist_Name'].values[0]))
```

```python
min_growth  = data[(data['Pop_2019'] / data['Pop_2009']) == min(data['Pop_2019'] / data['Pop_2009'])]
max_growth  = data[(data['Pop_2019'] / data['Pop_2009']) == max(data['Pop_2019'] / data['Pop_2009'])]
min_volati  = data[(data['Pop_2019'] - data['Pop_2009']) == min(data['Pop_2019'] - data['Pop_2009'])]
max_volati  = data[(data['Pop_2019'] - data['Pop_2009']) == max(data['Pop_2019'] - data['Pop_2009'])]
min_den2019 = data[(data['Pop_2019'] / data['Shape_Area']) == min(data['Pop_2019'] / data['Shape_Area'])]
max_den2019 = data[(data['Pop_2019'] / data['Shape_Area']) == max(data['Pop_2019'] / data['Shape_Area'])]
print('Phuong co toc do trang truong nhanh nhat  : {}'.format(max_growth['Com_Name'].values[0]))
print('Phuong co toc do tang truong thap nhat   : {}'.format(min_growth['Com_Name'].values[0]))
print('Phuong co bien dong dan so manh nhat {}'.format(max_volati['Com_Name'].values[0]))
print('Phuong co bien don dan so thap nhat : {}'.format(min_volati['Com_Name'].values[0]))
print('Phuong co mat do dan so cao nhat 2019   : {}'.format(max_den2019['Com_Name'].values[0]))
print('Phuong co mat do dan so thap nhat 21019   : {}'.format(min_den2019['Com_Name'].values[0]))
```
