```python
from sklearn.decomposition import PCA
import IPython
import matplotlib.pyplot as plt
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/lena.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
img = img.T
pca = PCA(n_components=10)
# nen anh
pca.fit(img)
img_reduced =pca.transform(img)
img_reduced = img_reduced.T 
cv2_imshow(img_reduced)
print(img_reduced.shape)
```

```python
# khoi phuc
img_recovered = pca.inverse_transform(img_reduced.T)
cv2_imshow(img_recovered.T)
print(img_recovered.T.shape)
```
